```python
def univalues(df):
    for column in df:
        if df[column].dtypes=='object':    
            print(f'{column} : {df[column].unique()}')
```

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
univalues(df1)
```

# Replacing Yes-No values with 1 and 0

```python
ync = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport',
      'StreamingTV','StreamingMovies','PaperlessBilling','Churn']

for col in ync:
    df1[col].replace({'Yes':1,'No':0}, inplace=True)
```

```python
for col in df1:
    print(f'{col}: {df1[col].unique()}')
```

```python
df1['gender'].replace({'Female':0,'Male':1},inplace=True)
```

```python
df1['gender'].unique()
```

```python
df2 = pd.get_dummies(data=df1,columns=['InternetService','PaymentMethod','Contract'])
df2.columns
```

```python
df2.dtypes
```

# Scaling Columns / One Hot Encoding

```python
cts = ['tenure','MonthlyCharges','TotalCharges'] 
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df2[cts] = scaler.fit_transform(df2[cts])
```

```python
df2.sample(2)
```

```python
for col in df2:
    print(f'{col}: {df2[col].unique()}')
```

```python
x = df2.drop('Churn', axis='columns')
y = df2['Churn']
```

# Training the model using ANN

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=5)
```

```python
len(x_train.columns)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),
    keras.layers.Dense(2000,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
    
])

model.compile(optimizer='adam',
             loss='binary_crossentropy',
             metrics = ['accuracy']
             
             )
model.fit(x_train,y_train,epochs=20)
```

```python
model.evaluate(x_test,y_test)
```

```python
yp = model.predict(x_test)
yp[1]
```

```python
ypred = [ ]
for element in yp:
    if element>0.5:
        ypred.append(1)
    else:
        ypred.append(0)
        
ypred
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test,ypred))
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=ypred)

plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
